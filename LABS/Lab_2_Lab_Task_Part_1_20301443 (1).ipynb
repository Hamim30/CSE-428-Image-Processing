```python
import numpy as np
from skimage import io, data, exposure
import matplotlib.pyplot as plt
```

# Task 1 (for bonus pts)

```python
# Import an image from skimage data
I = data.brick()
# io.imshow(I)
print(I.shape)
# Calculate the histogram of the image

def Hist(image):
    hist=[0]*256
    bins_hist=np.arange(0,256,1)
    for j in range(len(image)):
      for i in range(len(image)):
        hist[image[j, i]]  = hist[image[j, i]]+1


    return (hist, bins_hist)
```

```python

hist,bins_hist=(Hist(I))
plt.plot(bins_hist, hist)
plt.show()
```

# Task 2

```python
# Import the "good.jpeg" image
I = data.brick()

# normalize the image
I_norm=I/255

# perform arithmetic operations (contrast increase/decrease) on the image
I_add = np.clip(I_norm+0.5,0,1) # increase by 0.5
I_sub = np.clip(I_norm-0.4,0,1)# decrease by 0.4
I_mul =  np.clip(I_norm*2,0,1) # multiply by 2
I_div =  np.clip(I_norm/3,0,1)# divide by 3
```

# Task 3

```python
# plot in a (5, 2) grid

plt.figure(figsize=(10,25),dpi=75)
plt.subplot(5,2,1)
plt.title("I_norm Picture")
io.imshow(I_norm)
plt.subplot(5,2,2)
hist, bins_hist=exposure.histogram(I_norm,nbins=256, normalize=False)
plt.title("I_norm histogram")
plt.plot(bins_hist, hist)


plt.subplot(5,2,3)
plt.title("I_add Picture")
io.imshow(I_add)
plt.subplot(5,2,4)
plt.title("I_add Histogram")
hist, bins_hist=exposure.histogram(I_add,nbins=256, normalize=False)
plt.plot(bins_hist, hist)


plt.subplot(5,2,5)
plt.title("I_sub Picture")
io.imshow(I_sub)
plt.subplot(5,2,6)
plt.title("I_sub Histogram")
hist, bins_hist=exposure.histogram(I_sub,nbins=256, normalize=False)
plt.plot(bins_hist, hist)


plt.subplot(5,2,7)
plt.title("I_mul Picture")
io.imshow(I_mul)
plt.subplot(5,2,8)
plt.title("I_mul Histogram")
hist, bins_hist=exposure.histogram(I_mul,nbins=256, normalize=False)
plt.plot(bins_hist, hist)


plt.subplot(5,2,9)
plt.title("I_div Picture")
io.imshow(I_div)
plt.subplot(5,2,10)
plt.title("I_div Histogram")
hist, bins_hist=exposure.histogram(I_div,nbins=256, normalize=False)
plt.plot(bins_hist, hist)
```
